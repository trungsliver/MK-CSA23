# **Series**
- Khái niệm: là danh sách 1 chiều chứa dữ liệu, chỉ số phần tử (index) nằm ở danh sách khác. index có thể là bất kì kiểu dữ liệu nào trong numpy.

Ví dụ 1: index mặc định (không khai báo)
=> Series students chứa
  - Giá trị (value): Minh Anh, Lâm Khánh,...
  - Chỉ số phần tử (index): 0, 1, 2,..
  - Kiểu dữ liệu của series: object (tương ứng với str)

```python
students = pd.Series(['Minh Anh', 'Lâm Khánh', 'Tuấn Linh', 'Duy Anh'])
students
```

Ví dụ 2: có khai báo index, index có thể không phải kiểu int

```python
students = pd.Series(['Minh Anh', 'Lâm Khánh', 'Tuấn Linh', 'Duy Anh'],
                     index = ['A', 'B', 'C', 'D'])
students
```

Có thể tạo Series từ dictionary (kiểu object, giống json)

```python
fruits_dict = {
    'a': 'apple',
    'b': 'banana',
    'c': 'orange',
    'd': 'pear',
    'e': 'strawberry'
}
fruits = pd.Series(fruits_dict)
fruits
```

### Thực hành: Thống kê cân nặng học sinh
Cho các số liệu cân nặng học sinh như sau:
|Tên|Cân nặng|
|---|--------|
|Hoa|37 kg|
|Nam|48 kg|
|Hải|42 kg|
|Linh|39kg|
|Nhân|41kg|
|Phương|38kg|
|Thu|40 kg|
|Vũ|40 kg|
|Bảo|43 kg|

Sử dụng pandas Series, tìm tên các bạn tương ứng với các cân nặng `mean`, `median`, `mode`.

```python
# Chuyển data từ đề bài thành dạng Series
weight_dict = {
    'Hoa': 37,
    'Nam': 48,
    'Hải': 42,
    'Linh': 39,
    'Nhân': 41,
    'Phương': 38,
    'Thu': 40,
    'Vũ': 40,
    'Bảo': 43
}
# Tạo series
weights = pd.Series(weight_dict)
weights
```

```python
# Tìm tên tương ứng với mean, median, mode
    # mean (trung bình): tổng tất cả các giá trị chia cho số lượng giá trị
mean_weight = weights.mean()
    # median (trung vị): giá trị ở giữa khi series đã được sắp xếp theo giá trị
median_weight = weights.median()
    # mode: giá trị xuất hiện nhiều lần nhất
mode_weight = weights.mode()

# Hiển thị kết quả
print('Cân nặng trung bình:', mean_weight, 'kg')
print('Cân nặng trung vị:', median_weight, 'kg')
print('Cân nặng xuất hiện nhiều lần nhất:', mode_weight.values)
print('Cân nặng xuất hiện nhiều lần nhất:', end = ' ')
for i in mode_weight:
    print(i)
```

**Thực hành 2: Đánh giá chất lượng dịch vụ**

```python
ratings = pd.Series([4,3,1,2,1,5,4,1,1])
# Đánh giá
print(f'Chất lượng trung bình: {round(ratings.mean(),2)}/5')
print(f'Đánh giá phổ biến nhất: {ratings.mode().values[0]} sao')
```

# **DataFrame**
- Khái niệm: là cấu trúc dữ liệu dạng 2 chiều (bảng, gồm hàng và cột)
- Mỗi cột trong DataFrame là 1 series

```python
# keys: tương ứng với tên cột
# danh sách values: tương ứng với giá trị cột đó
students = pd.DataFrame({
    'name': ['Minh Anh', 'Lâm Khánh', 'Tuấn Linh', 'Duy Anh', 'Thu Hương'],
    'age': [15, 15, 16, 17, 17],
    'gender': ['male', 'male', 'male', 'male', 'female']
})
students
```

Tóm tắt thông tin DataFrame: DataFrame.info()
- RangeIndex: số bản ghi (số dòng)
- Data Colums: số cột (số Series có trong DataFrame)


```python
students.info()
```

DataFrame.head(n): Hiển thị n giá trị đầu của DataFrame

```python
students.head(3)
```

DataFrame.tail(n): Hiển thị n giá trị cuối của DataFrame

```python
students.tail(3)
```

**Bài tập 1:**
Tạo một pandas DataFrame từ dữ liệu sau:
| Name | Age | Score |
| ---- | --- | ----- |
| An   | 14  | 8     |
| Bình | 15  | 7     |
| Lan  | 14  | 9     |

Yêu cầu:
- Import thư viện pandas
- Tạo DataFrame từ dữ liệu trên
- In toàn bộ bảng dữ liệu
- In riêng cột Score


```python
data = {
    "Name": ["An", "Bình", "Lan"],
    "Age": [14, 15, 14],
    "Score": [8, 7, 9]
}

df = pd.DataFrame(data)

# In toàn bộ DataFrame
print('DataFrame:')
print(df)

# In cột Score
print('\nScore:')
print(df["Score"])
```

**Bài tập 2:**
| Name | Score |
| ---- | ----- |
| An   | 8     |
| Bình | 7     |
| Lan  | 9     |
| Mai  | 6     |
| Nam  | 8     |

Yêu cầu:
- Tạo DataFrame từ dữ liệu trên
- Tính: điểm trung bình (mean), điểm cao nhất (max), thấp nhất (min), điểm xuất hiện nhiều nhất (mode)
- In kết quả ra màn hình

```python
data = {
    "Name": ["An", "Bình", "Lan", "Mai", "Nam"],
    "Score": [8, 7, 9, 6, 8]
}

df = pd.DataFrame(data)

mean_score = df["Score"].mean()
max_score = df["Score"].max()
min_score = df["Score"].min()
mode_score = df["Score"].mode()

print("Mean:", mean_score)
print("Max:", max_score)
print("Min:", min_score)
print("Mode:", mode_score.values[0])
```
